```python
import zipfile
import os

zip_path = "/content/IR motorcycle images.zip"
extract_path = "/content/dataset"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print("Unzip done!")
```

```python

import os

dataset_path = "/content/dataset"   # change if your folder name is different

image_extensions = (".jpg", ".jpeg", ".png", ".bmp", ".tiff", ".webp")

count = 0

for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.lower().endswith(image_extensions):
            count += 1

print("Total images in dataset:", count)
```

```python
import cv2
import numpy as np
import os

input_root = "/content/dataset"      # where zip was extracted
output_folder = "/content/enhanced_output"
os.makedirs(output_folder, exist_ok=True)

def enhance_image(img):
    # Gamma correction
    gamma = 0.4
    lookUpTable = np.array([((i / 255.0) ** gamma) * 255 for i in range(256)]).astype("uint8")
    gamma_corrected = cv2.LUT(img, lookUpTable)

    # CLAHE
    clahe = cv2.createCLAHE(clipLimit=4.0, tileGridSize=(8,8))
    contrast = clahe.apply(gamma_corrected)

    # Sharpen
    kernel = np.array([[0,-1,0],
                       [-1,5,-1],
                       [0,-1,0]])
    sharpened = cv2.filter2D(contrast, -1, kernel)

    return sharpened


count = 0

# Walk through all folders and subfolders
for root, dirs, files in os.walk(input_root):
    for file in files:
        if file.lower().endswith((".jpg", ".png", ".jpeg")):
            img_path = os.path.join(root, file)

            img = cv2.imread(img_path, 0)
            if img is None:
                continue

            enhanced = enhance_image(img)

            save_name = f"img_{count}.png"
            cv2.imwrite(os.path.join(output_folder, save_name), enhanced)

            count += 1

print(f"✅ Done! {count} images saved to {output_folder}")
```

```python
import shutil

folder_to_zip = "/content/enhanced_output"   # change if your folder name is different
zip_path = "/content/dataset.zip"

shutil.make_archive("/content/enhanced_output", 'zip', folder_to_zip)

print("Zipped successfully!")
```
